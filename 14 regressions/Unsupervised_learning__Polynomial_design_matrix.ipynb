```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n  = 30
x  = np.linspace(-2,4,n)
y1 = x**2 + np.random.randn(n)
y2 = x**3 + np.random.randn(n)
```

```python
X1 = np.array([np.ones(n), x, x**2]).T
beta1 = np.linalg.inv(X1.T @ X1) @ X1.T @ y1.T
y1hat = X1@beta1

X2 = np.array([np.ones(n), x, x**2, x**3]).T
beta2 = np.linalg.inv(X2.T @ X2) @ X2.T @ y2.T
y2hat = X2@beta2
```

```python
plt.plot(x,y1,'ko',markerfacecolor='r',label='y1')
plt.plot(x,y2,'ko',markerfacecolor='g',label='y2')
plt.plot(x,y1hat,'r',label='y1 fit')
plt.plot(x,y2hat,'g',label='y2 fit')
plt.legend()
plt.show()
```
