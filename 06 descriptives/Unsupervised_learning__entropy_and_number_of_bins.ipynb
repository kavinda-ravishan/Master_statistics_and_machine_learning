```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 1000
brownnoise = np.cumsum( np.sign(np.random.randn(N)) )
```

```python
i = 1
start = 4
stop = 50
step = 2

entro = np.zeros(int(((stop-start)/step)))
bin = np.zeros(int(((stop-start)/step)))

i = 0
for nbins in range(start,stop,step):
    # bin the data and convert to probability
    nPerBin,bins = np.histogram(brownnoise,nbins)
    probs = nPerBin / sum(nPerBin)

    # compute entropy
    entro[i] = -sum( probs*np.log2(probs+np.finfo(float).eps) )
    bin[i] = nbins
    i += 1
```

```python
plt.plot(bin, entro, '-s')
plt.xlabel('Bins')
plt.ylabel('Entropy')
plt.title('No of Bins Vs Entropy')
plt.show()
```
