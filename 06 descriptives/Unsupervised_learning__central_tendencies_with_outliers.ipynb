```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N1 = 1000
N2 = int(N1*0.1)
data1 = np.random.randn(N1)
```

```python
data2 = 3*np.random.randn(N2)+80

y1,x1 = np.histogram(data1, 10)
x1 = (x1[:-1]+x1[1:])/2

y2,x2 = np.histogram(data2, 10)
x2 = (x2[:-1]+x2[1:])/2

data = np.hstack((data1,data2))
dataMean = np.mean(data)
dataMedian = np.median(data)

plt.plot(x1,y1,'b',x2,y2,'b')
plt.plot([dataMean, dataMean], [0, max(y1)], 'b--')
plt.plot([dataMedian, dataMedian], [0, max(y1)], 'r--')
plt.xlim([-10, 100])
plt.show()
```
