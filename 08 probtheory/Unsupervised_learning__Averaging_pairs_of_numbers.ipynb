```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
udist1 = (99*np.random.rand(N,1))+1
udist2 = (99*np.random.rand(N,1))+1
```

```python
avg = np.zeros(N*N)

n = 0
for i in range(N):
    for j in range(N):
        avg[n] = (udist1[i]+udist2[j])/2
        n=n+1
```

```python
plt.hist(avg)
plt.show()
```
