```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
# the null-hypothesis value
H0val = 0
N = 20 # sample size
popMu = .5 # true population mean
stdev = np.linspace(0.1, 3, 100) #  standard deviation
p = np.zeros(stdev.shape[0], dtype=np.float64)
tval = np.zeros(stdev.shape[0], dtype=np.float64)
```

```python
for i in range(0,stdev.shape[0]):
    data = stdev[i]*np.random.randn(N) + popMu
    tval[i],p[i] = stats.ttest_1samp(data,H0val)
```

```python
plt.plot(stdev, tval, 'bs')
plt.title('T Vs stdev')
plt.xlabel('standard deviation')
plt.ylabel('T-Value')
plt.show()
```

```python
plt.plot(stdev, p, 'bs')
plt.title('P Vs stdev')
plt.xlabel('standard deviation')
plt.ylabel('Probability')
plt.show()
```

```python
plt.plot(tval, p, 'bs')
plt.title('T Vs P')
plt.xlabel('T-Value')
plt.ylabel('Probability')
plt.show()
```
