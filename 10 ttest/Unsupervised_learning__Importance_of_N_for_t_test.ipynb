```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
mean1 = 1.2
mean2 = 1
pval_thresh = 0.05

N = np.arange(10,201)
num_exps = 100
```

```python
tstats = np.zeros([num_exps, len(N)])
tval_thresh = np.zeros(len(N))
```

```python
for e in range(0, num_exps):
    for i in range(len(N)):
        data1 = mean1 + np.random.randn(N[i])
        data2 = mean2 + np.random.randn(N[i])

        tstats[e, i], _ = stats.ttest_ind(data1, data2, equal_var=True)
```

```python
for e in range(0, num_exps):
    plt.plot(N, tstats[e, :], color=[0.8, 0.8, 0.8])

plt.plot(N, np.mean(tstats, axis=0), 'black')
plt.xlabel('Sample size')
plt.ylabel('T-test value')
plt.show()
```
