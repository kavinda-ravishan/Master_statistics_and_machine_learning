```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
N = 40

docuRatings = np.random.randint(1, 5, N)

noice = np.arange(4, 100, 1)
noExp = 100

crk = np.zeros((noExp, noice.shape[0]))
crp = np.zeros((noExp, noice.shape[0]))
```

```python
for i in range(noice.shape[0]):
    for j in range(noExp):
        eduLevel = np.ceil( (docuRatings + np.random.randint(1, noice[i], N) )/(noice[i] + 5) * 4 )

        crk[j,i] = stats.kendalltau(eduLevel,docuRatings)[0]
        crp[j,i] = stats.pearsonr(eduLevel,docuRatings)[0]
```

```python
plt.plot(np.abs(np.mean(crk, 0) - np.mean(crp, 0)), 'r')
plt.xlabel('noice')
plt.ylabel('correlation difference')
plt.grid()
```
