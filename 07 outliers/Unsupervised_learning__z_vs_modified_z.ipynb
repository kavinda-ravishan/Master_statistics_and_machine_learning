```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels import robust
import scipy.stats as stats
```

```python
zscorethresh = 3
N = 400
```

```python
data = np.random.rand(N)**4

mean = np.mean(data)
std = np.std(data)
dataZ = (data - mean)/std

dataMed = np.median(data)
dataMAD = robust.mad(data)
norm = stats.norm.ppf(.75)
dataMz = (norm*(data - dataMed))/dataMAD

outliersZ = np.where(np.abs(dataZ)>zscorethresh)[0]
outliersMz = np.where(np.abs(dataMz)>zscorethresh)[0]
```

```python
fig,ax = plt.subplots(2,1,figsize=(10,10))

ax[0].plot(dataZ,'^k',markerfacecolor='w',markersize=10)
ax[0].plot([0,N],[zscorethresh,zscorethresh],'r--',linewidth=3)
ax[0].plot([0,N],[-zscorethresh,-zscorethresh],'r--',linewidth=3)
ax[0].plot(outliersZ,dataZ[outliersZ],'x',color='r',markersize=10)
ax[0].grid(color='black', linestyle='--', linewidth=1)
ax[0].set_title('z-score')

ax[1].plot(dataMz,'^k',markerfacecolor='w',markersize=10)
ax[1].plot([0,N],[zscorethresh,zscorethresh],'r--',linewidth=3)
ax[1].plot([0,N],[-zscorethresh,-zscorethresh],'r--',linewidth=3)
ax[1].plot(outliersMz,dataMz[outliersMz],'x',color='r',markersize=10)
ax[1].grid(color='black', linestyle='--', linewidth=1)
ax[1].set_title('special z-score')

plt.show()
```
