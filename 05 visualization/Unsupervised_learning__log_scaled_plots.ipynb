```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = [n/5 for n in range(0,25)]
y1 = x[:]
y2 = [math.exp(n) for n in x]
```

```python
plt.plot(x,y1,'s-',label='y = x')
plt.plot(x,y2,'o-',label='y = e^x')

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
plt.semilogy(x,y1,'s-',label='y = x')
plt.semilogy(x,y2,'o-',label='y = e^x')

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
