```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
## create data for the histogram

# number of data points
n = 1000

# generate data - log-normal distribution
data = np.exp( np.random.randn(n)/2 )
```

```python
plt.plot(data,'s')
plt.show()
```

```python
# show as a histogram

# number of histogram bins
k = 40

plt.hist(data,bins=k)
plt.show()
```

```python
# another option
y,x = np.histogram(data,bins=k)

# bin centers
# x = [0, 1, 2, 3]
# x[ :-1] : [0, 1, 2]
# x[1:  ] : [1, 2, 3]
# x[ :-1] + x[1:  ] = [ 0+1, 1+2, 2+3 ]
xx = (x[1:]+x[:-1])/2

plt.plot(xx,y)
plt.show()
```
